```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
df_rec=pd.read_excel('meal.xlsx')
df_rec.head()
```

```python
df_user=pd.read_excel('meal.xlsx')
df_user.head()
```

```python
#for list 1
df_user['order_list1'] = df_user['order_list1'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df_user.head()
```

```python
df_user['order_list1']=df_user['order_list1'].apply(lambda x:"".join(x))
```

The above 2 functions have been done to combine the words into one word and will be done for all the 4 remaining order lists

```python
#for list 2
df_user['order_list2'] = df_user['order_list2'].apply(lambda x:[i.replace(" ","") for i in x])
df_user['order_list2']=df_user['order_list2'].apply(lambda x:"".join(x))
df_user.head()
```

```python
#for list 3
df_user['order_list3'] = df_user['order_list3'].apply(lambda x:[i.replace(" ","") for i in x])
df_user['order_list3']=df_user['order_list3'].apply(lambda x:"".join(x))
df_user.head()
```

```python
#for list 4
df_user['order_list4'] = df_user['order_list4'].apply(lambda x:[i.replace(" ","") for i in x])
df_user['order_list4']=df_user['order_list4'].apply(lambda x:"".join(x))
df_user.head()
```

```python
#for list 5
df_user['order_list5'] = df_user['order_list5'].apply(lambda x:[i.replace(" ","") for i in x])
df_user['order_list5']=df_user['order_list5'].apply(lambda x:"".join(x))
df_user.head(10)
```

```python
df_user['tags']=df_user['order_list1']+" "+df_user['order_list2']+" "+df_user['order_list3']+" "+df_user['order_list4']+" "+df_user['order_list5']
```

```python
new_df=df_user[['User_id','Order_id','order_list5','tags']]
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv =CountVectorizer(max_features=500,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[10])),reverse=True,key=lambda x:x[1])
# sorts on the basis of the similarity using the labda function
```
